```python
from io import StringIO
```

```python
inp = StringIO('''0 <-> 2
1 <-> 1
2 <-> 0, 3, 4
3 <-> 2, 4
4 <-> 2, 3, 6
5 <-> 6
6 <-> 4, 5''')


inp = open('12.inp')
instructions = []
for instr in inp.readlines():
    s = instr.strip().split()
    instructions = instructions + [[s[0], s[2:]]]
    
instructions
```

```python
places = {}

for i in instructions:
    if not i[0] in places:
        places[i[0]] = {'links':None}
    for j in i[1]:
        if places[i[0]]['links'] is None:
            places[i[0]]['links'] = [j.strip(',')]
        else:
            places[i[0]]['links'] += [j.strip(',')]
        

places
    
```

```python
number = 0
not_explored = {}
for i in range(1999):
    not_explored[str(i)] = 1
    


def explore(place):
    global number
    number += 1
    del not_explored[place]
    for link in places[place]['links']:
        if link in not_explored:
            explore(link)
    
```

```python
groups = 0



while not_explored:
    start = list(not_explored.keys())[0]
    global groups
    groups += 1
    explore(start)
    print(number)
    
```
