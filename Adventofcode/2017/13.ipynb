```python
strinp = StringIO('''0: 3
1: 2
2: 4
4: 4
6: 5
8: 8
10: 6
12: 6
14: 8
16: 6
18: 6
20: 8
22: 12
24: 8
26: 8
28: 12
30: 8
32: 12
34: 9
36: 14
38: 12
40: 12
42: 12
44: 14
46: 14
48: 10
50: 14
52: 12
54: 14
56: 12
58: 17
60: 10
64: 14
66: 14
68: 12
70: 12
72: 18
74: 14
78: 14
82: 14
84: 24
86: 14
94: 14''')

#strinp = StringIO('''0: 3
#1: 2
#4: 4
#6: 4''')

inp = []
for i in strinp.readlines():
    l = [int(x) for x in (i.strip('').strip('\n').split(':'))]
    inp.append(l)
    
```

```python
class Layer:
    
    def __init__(self, length):
        self.pos = 0
        self.direction = 1
        self.length = length
        
    def tick(self):
        if self.pos == self.length - 1:
            self.direction = -1
            self.pos = self.pos + self.direction
        elif self.pos == 0:
            self.direction = 1
            self.pos += self.direction
        else:
            self.pos = self.pos + self.direction
            
    def print_pos(self):
        print(self.pos)
        
    def get_pos(self):
        return self.pos
        
```

```python
 
layers = {}
for i in inp:
    layers[i[0]] = Layer(i[1])
    
print (layers)

def tick_layers():
    for l in layers:
        layers[l].tick()
        
def print_layers():
    for k in layers:
        print ("key: {}, pos: {}".format(k, layers[k].get_pos()))
        
pos = 0
hit = 0
for p in range(100):
    print ("entering layer: {}".format(p))
    print_layers()
    if p in layers and layers[p].get_pos() == 0:
        hit += p * layers[p].length
        print("hit!")
    tick_layers()
        
        
print (hit)

```
