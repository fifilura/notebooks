```python
from io import StringIO
```

```python
test = StringIO('''set a 1
add a 2
mul a a
mod a 5
snd a
set a 0
rcv a
jgz a -1
set a 1
jgz a -2''')

test = open('18.inp')

commands = []


for line in test.readlines():
    line = line.replace('\n','')
    spl = line.split(' ')
    command = {}
    command['c'] = spl[0]
    command['s'] = spl[1]
    if len(spl) == 3:
        command['p'] = spl[2]
    commands = commands + [command]
```

```python
registers = {}
from string import ascii_lowercase
for c in ascii_lowercase:
    registers[c] = 0
registers
```

```python
commands
```

```python
def get_val(val):
    if val in ascii_lowercase:
        return registers[val]
    return int(val)
```

```python
pc = 0


last_played_sound = 0

while True:
    c = commands[pc]['c']
    s = commands[pc]['s']
    
    if c == 'set':
        registers[s] = get_val(commands[pc]['p'])
    elif c == 'add':
        registers[s] = registers[s] + get_val(commands[pc]['p'])
    elif c == 'mul':
        registers[s] = registers[s] * get_val(commands[pc]['p'])
    elif c == 'mod':
        registers[s] = registers[s] % get_val(commands[pc]['p'])
    elif c == 'snd':
        print("playing sound: {}".format(registers[s]))
        last_played_sound = registers[s]
    elif c == 'rcv':
        if get_val(s) != 0:
            print ('last_played_sound: {}'.format(last_played_sound))
            break
    elif c == 'jgz':
        if get_val(s) > 0:
            pc += get_val(commands[pc]['p'])
            continue
    else:
        print ("Not caught")
    pc += 1
```
