```python
def calc_tot_weight(node):
    if 'children' in node:
        print (node['children'])
        node['tot_weight'] = node['weight'] + sum(map(calc_tot_weight,  [nodes[x] for x in node['children']]))
    else:
        node['tot_weight'] = node['weight']
    return node['tot_weight']
    
calc_tot_weight(nodes[top])
```
