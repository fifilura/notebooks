```python
inpf = open('1.txt')
inp = []
for i in inpf.readlines():
    inp = inp + i.strip().split(',')
    
```

```python

res = 0
res_list = {0}
while True:
    for i in inp:
        res = res + int(i)
        if res in res_list:
            print(res)
            break
        res_list.add(res)

    
print(res)
    
```
