```python
class Node:
    def __init__(self, value):
        self.pred = None
        self.suc = None
        self.value = value
    
    def removeme(self):
        if self.pred:
            self.pred.suc = self.suc
        if self.suc:
            self.suc.pred = self.pred
        
    def printme(self):
        print("pred: {} me: {} suc: {}".format(self.pred.value if self.pred else "NULL", self.value, 
                                               self.suc.value if self.suc else "NULL"))
        
    
# insert
pred = None
first_node = None
for c in exp:
    n = Node(c)
    if not first_node:
        first_node = n
    n.pred = pred
    if n.pred:
        n.pred.suc = n
    pred = n
        

cur = first_node
prev_val = None


def get_list():
    res = ''
    cur = first_node
    while cur:
        res += cur.value
        cur = cur.suc
    return res


while cur:
    print(cur.value)
    print("cur_val: {} prev_val: {}".format(cur.value, prev_val))
    if prev_val and np.abs(ord(cur.value) - ord(prev_val)) == 32:
        print("collapse")
        print_list()
        print(cur.pred.value)
        cur.printme()
        cur.pred.removeme()
        cur.removeme()
        if cur.pred:
            cur = cur.pred
        else:
            first_value = cur.suc
            cur = cur.suc
        

        
    prev_val = cur.value
    cur = cur.suc
print(get_list())
```
