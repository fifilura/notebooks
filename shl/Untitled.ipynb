```python
df = pd.read_csv('2017_2018.csv')


def calc_full_time_score(period_data, home, number_of_periods):

    if period_data != period_data:
        return None
    
    periods = period_data.split(",")[0:number_of_periods]
    score = 0
    index = 1
    
    if home:
        index = 0
        
    for p in periods:
        score += int(p.strip().split("-")[index])
        
    return score
    
df['full_time_score'] = df['PERIOD'].apply(lambda x: calc_full_time_score(x, True, 3))
df
```
